## 1.0 Data Preparation and Cleaning

```
# We first start by importin the pandas library
# 
import pandas as pd
```

## 1.1 Loading our data and creating datasets from the files provided

```
#loading the cells_geo dataset and creating a dataframe

dfcells_geo = pd.read_excel('cells_geo.xlsx')

#Previewing the dataset to see what it looks like
dfcells_geo
```

```
#loading the Telcom_dataset3 dataset and creating a dataframe

dfTelcom_day3 = pd.read_csv('Telcom_dataset3.csv')

#Previewing the dataset to see what it looks like
dfTelcom_day3
```

```
#loading the Telcom_dataset2 dataset and creating a dataframe

dfTelcom_day2 = pd.read_csv('Telcom_dataset2.csv')

#Previewing the dataset to see what it looks like
dfTelcom_day2
```

```
#loading the Telcom_dataset dataset and creating a dataframe

dfTelcom_day1 = pd.read_csv('Telcom_dataset.csv')

#Previewing the dataset to see what it looks like
dfTelcom_day1
```

## 1.2 Splitting the column in geo_cells to respective number of columns with the correct column names


```
#Splitting the column in geo_cells to respective number of columns in the dataset seperated by ';' or ';;'

dfcells_geo1 = dfcells_geo[';VILLES;STATUS;LOCALISATION;DECOUPZONE;ZONENAME;LONGITUDE;LATITUDE;REGION;AREA;CELL_ID;SITE_CODE'].apply(lambda x: pd.Series([i for i in x.lower().split(';'or';;')]))



#Previewing the new columns
dfcells_geo1

#Renaming the new columns with their respective names
dfcells_geo2 = dfcells_geo1.rename(columns = {1:'CITY',2:'STATUS', 3:'LOCALISATION', 4:'DECOUPZONE', 5:'ZONENAME', 6:'LONGITUDE', 7:'LATITUDE', 8:'REGION', 9:'AREA', 10:'CELL_ID', 11:'SITE_CODE'}) 
dfcells_geo2

#dropping the column zero which is exactly the same as the index generated by pandas
dfcells_geo3 = dfcells_geo2.drop([0], axis = 1)
dfcells_geo3
```

## 1.3 We now merge TelecomDataset, TelecomDataset2 and TelecomDataset3 for future use


```
#We now merge TelecomDataset, TelecomDataset2 and TelecomDataset3 for future use
#This is because the shapes of the datasets are identical

#We first rename the column PRODUTC to PRODUCT and DATETIME to DATE_TIME in TelecomDataset before merging and previewing it
dfTelcom_day1 = dfTelcom_day1.rename(columns = {'PRODUTC':'PRODUCT', 'DATETIME':'DATE_TIME'})
dfTelcom_day1

#We rename DW_B_NUMBER & DW_A_NUMBER to DW_B_NUMBER_INT & DW_A_NUMBER_INT in Telcom_dataset2 and previewing it
dfTelcom_day2 = dfTelcom_day2.rename(columns = {'DW_B_NUMBER':'DW_B_NUMBER_INT', 'DW_A_NUMBER':'DW_A_NUMBER_INT' })
dfTelcom_day2

#We rename CELLID & SIET_ID to CELL_ID & SITE_ID in Telcom_dataset3 and previewing it
dfTelcom_day3 = dfTelcom_day3.rename(columns = {'CELLID':'CELL_ID', 'SIET_ID':'SITE_ID' })
dfTelcom_day3

#We then merge the first two dataframes for TelecomDataset, TelecomDataset2 created earlier
df_twodays = pd.concat((dfTelcom_day1, dfTelcom_day2), axis=0, ignore_index=True)
df_twodays

#We now merge the two dataframes with the third one, TelecomDataset3
df_threedays = pd.concat((df_twodays, dfTelcom_day3), axis=0, ignore_index=True)
df_threedays
```

## 1.4 We now merge df_threedays DataFrame with dfcells_geo3 DataFrame

```
#We rename SITE_CODE to SITE_ID in Telcom_dataset3 dfcells_geo3 previewing it
dfcells_geo3 = dfcells_geo3.rename(columns = {'SITE_CODE':'SITE_ID'})
dfcells_geo3

#We now merge the DataFrame df_threedays with dfcells_geo3 DataFrame 
df_main = pd.concat((df_threedays, dfcells_geo3), axis=0, ignore_index=True)
df_main

#From the dataset we see there are a lot of missing values so we cant do away with them because we will lose 15,003 - 3974 = 11, 029 rows of data
#We will just remove the duplicates in the DataFrame and replace Nan values with zero
df_main1 = df_main.drop_duplicates()
df_main2 = df_main1.fillna(0)
df_main2
```
